## Obvious oversights

### Documentation

Currently, documentation is somewhat hard to find. There is useful data in the FAQ and Resources sections, but it's not obvious from inside the software. The main app probably doesn't need it, but all that information should be at the fingertips of someone in the Sandbox and in some cases inside the app.

The Sandbox should have the FAQ in it, or relevant subset thereof; one of the draggable panels could contain an accordion, perhaps. Merge FAQ and Resources in Sandbox, with relevant FAQ entries containing the links to the PDFs so it's as upfront as possible.

We have the Comatose help popups in some places; they could be more detailed as they only show when asked and so don't get in the way. They could also be more numerous.

### Links to YouTube videos

Get either Pete or Lewis to do a full, detailed demo (max 20 mins - look into getting one of them a copy of ScreenFlow) of each of the tools - Innovation, Market, Category and SIM - in a fullscreen browser session. Record the screen and voice, ideally without mouse cursor, clean up with Final Cut (noise etc), post to YouTube as updated tutorial sessions. Link the videos in the signup emails for each tool, and also on the documentation page.

### Tutorial wizard [USER STORY]

Scenario: 
 - User signs up, logs in. 
 - First thing they see is a center-screen overlay panel with a message something akin to "We see you're new here; would you like to go to the sandbox and take a tutorial on setting up an event?"
 - Click 'No thanks, I've used the software before' and it goes away.
 - Click 'Yes please, show me' and it links to the sandbox and immediately starts a click-by-click tutorial.
  - Overlay panel at top of screen with instructions ("[...]")
  - Arrow used to direct user attention
  - Arrow points to existing simple lot in Draft Lots panel [This is a draft Simple Lot we have created for you. Click on the name to see the Lot details]
   - ........
  - Arrow points to existing advanced lot in Draft Lots panel [This is a draft Advanced Lot we have created for you. Click on the name to see the Lot details]
   - {user clicks}
   - {arrow takes user through Lot Settings Tab, then [click 'Save and go to next step' when you're ready]}
   - Draft lot has one of each input type (H/P P/T/N/D/K = 10 Lot Components), plus an extra PT.
    - Arrow points to the dustbin on the extra PT [We don't need this Lot Component. Click here to remove it.]
    - {user clicks}
    - Arrow points to pencil icon on Host Price cell [This is a Host Price. You can set a Price here, for example to (why would anyone actually do this?). Click here to adjust the price details.]
    - {user clicks}
    - Arrow points to 'Save' button. [You can set Current Price (useful for you to know how much you're saving), Qualification price (...) etc etc. When you're done, just click 'Save' to set the Lot Component details.
    - {user clicks}
    - Same again with a participant field. [With participant fields, you can only set the Description. This is helpful for the participants to know what you want, in more detail than the Lot Component title might provide.]
    - {user clicks, sets description, clicks 'save'}
    - Arrow points to 'Save and Go To Next Step'. [Click here when you're satisfied you know how a Lot is constructed]
    - {user clicks}
    - Arrow points to name of pre-populated document. [This is an attached document. Click here to view (etc)]
    - {stuff}
    - {user clicks 'Save and Go To Next Step'}
    - One participant is already populated. [We have added one participant already, but one bidder does not an auction make. You'll need some more.]
    - {short time passes}
    - Arrow points to 'Add New Participants'. [You can invite participants to bid on your events here]
    - {user clicks}
    - Arrows point to checkboxes next to suppliers 2 and 3. [Check the remaining Sandbox Suppliers to add them to the event]
    - {both checkboxes are set}
    - Arrow points to 'Add'. [Click here to save the new suppliers]
    - {user clicks}
    - Arrow points to 'Automatically accept new participants'. [You can choose to approve every supplier personally, or just have everyone with an invite code automatically accepted. Select this for now.]

    - Arrow points to 'Invite Participants and Finish'. [Excellent. Now we can start the event.]

And so it goes on, including bidding as a participant and a 'halt this event' at the end. When the event is finished, download reports.
    

### Popups on login

Or 'Hints and Tips', perhaps. The emails sent out with hints would be a good start for content. A lot of software when first run pops up a 'hints' box, which can be stepped through, dismissed, or permanently dismissed. This would be helpful the first few times, especially in combination with an optional click-by-click tutorial.
