**Award Lots**

 - Requires custom reports
 - **User Story stage 1:** 
  - 1) Add an 'Include in Award Notice' column with an info-help icon
  - 2) It defaults to green tick = Yes when an 'Awarded Participant' is selected from the drop-down
  - 3) User can toggle to red cross = No
  - 4) No tick or cross option if 'None' is selected in 'Awarded Participant' column
  - 5) Soft blue banner text at the top needs to change to, "Use this pop-up to record how you intend to award the Lots to your participants.  You can also notify them via the tool by choosing who to notify and creating a new award notice to send them.  You can do this in stages if you wish to bespoke each notification."
  - 6) A new button called 'Create New Award Notice' is added in that adds more options to the bottom of the pop-up (see 'adding scorers' pop-up for a working example)
  - 7) Clicking 'Create New Award Notice' shows error message if no Lot has been toggled to 'Include in Award Notice', "You have not selected any Lots to include in this notice."
  
 - **User Story stage 2:**
  - 1) Clicking the 'Create New Award Notice' button makes extra fields reveal, much like we do on 'Add New Scorer' pop-up.
  - 2) A text field to enter a Purchase Order Number.  If used, it includes the number in the Event Report (excel) and also in the Award Notice Email, just as an extra row of text.
  - 3) An expandable text field to edit the notification email text, with a confirmation of that text to the right to show what you're writing.  Please consult 'Approve Decision' pop-up in SIM for an example.
  - 4) Click 'Send Award Notice' to send out the email to the Awarded Participants.  Send 1 email to each Participant, listing out the Lots that participant has been awarded.  The email will contain a Bid Report for the Participant detailing the Lots that they have been awarded (see other user story).
  - 5) Once sent, hide all the New Award Notice details to return to the pop-up as it was, though with new details; See below.
  - 6) Show a confirmation message saying "The Award Notice has been sent.  Your Award details have been updated to confirm this."
  
 - **User Story stage 3:**
  - 1) It marks which Lot has been notified using a small green tick and the text 'Notified', using the same style we see on 'Status' column on an event Participant tab.
  - 2) It also adds a comment icon next to it so you can see the notification text that went with this.  
  - 3) Clicking the comment box will bring up a pop-up/info box of when the notification was made, what message was used and what PO was entered.
  - 4) Further notifications can be made after this, but only to Lots that haven't been notified already.
  - 5) Lots that have been notified will not receive further notification messages and will be locked down.  The Participant drop-down will be greyed out.
  - 6) The Host 'Download Event Report' excel report will contain the information in the 'Award Lots' tab.  Please see the top of this page for details.
